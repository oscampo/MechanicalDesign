Se definen valores iniciales de longitud y de valores para la coordenada $x$ a lo largo de la longitud del elemento.  Las reacciones serán $Ay$ y $By$.  $Lx$ es el valor de frontera para el cual $x=L^+$

```python
L=25.0
x=linspace(0,L,100) #Se define un vector x con 100 valores igualmente espaciados entre 0 y L
Ay=Symbol('Ay')
By=Symbol('By')
X=Symbol('X')
Lx=L+pow(10,-10) # Esto es L+10^-10
```

Se define la función de cargas, teniendo en cuenta que se tiene una carga distribuida que empieza en $x=5cm$ y termina (se anula) en $x=15cm$.  Es decir, matricialmente:

```python
qm=array([[Ay,-50,50,By],[0,5,15,L],[-1,0,0,-1]]) #Definición de una matriz
qm
```

Lo cual en notación clásica resulta:

```python
qx=mat2fsin(qm,X) # la función mat2fsin() evalúa un modelo de funciones de singularidad expresado 
                  # en forma matricial. Si el punto de evaluación es una variable simbólica, entonces 
                  # devuelve el equivalente algebráico del modelo
qx
```

De la expresión matricial de $q(x)$ se obtiene $V(x)$ y $M(x)$

```python
Vm=ifsin(qm) #La función ifsin() integra un modelo de funciones de singularidad expresado matricialmente
Mm=ifsin(Vm)
print('V(x)=\n',Vm,'\n M(x)=\n',Mm)
```

Lo cual, evaluado en $x=L^+$, resulta:

```python
V=mat2fsin(Vm,Lx)
M=mat2fsin(Mm,Lx)
print('V(x)=\n',V,'\n M(x)=\n',M)
```

Al resolver este sistema, se obtiene:

```python
r=solve([V,M])
Ay=float(r[Ay])
By=float(r[By])
r
```

Reemplazamos los valores encontrados:

```python
qm=array([[Ay,-50,50,By],[0,5,15,L],[-1,0,0,-1]])
Vm=ifsin(qm)
Mm=ifsin(Vm)
```

Ahora se evalúa $V(x)$ y $M(x)$ en cada uno de los valores del vector $x$

```python
Mx=mat2fsin(Mm,x)
Vx=mat2fsin(Vm,x)
```

Con lo anterior hemos creado vectores que definen el cortante y el momento del elemento. Del vector de momentos hallamos la ubicación y valor de $Mmax$

```python
Mmax=max(Mx)
ubicacion=x[where(Mx==Mmax)[0]]
print('Mmax=', Mmax, 'N/cm @', ubicacion, 'cm')
```

Ahora podemos graficar el cortante y el momento

```python
from pylab import *
figure()
subplot(2,1,1)
fill_between(x,Mx,x*0,alpha=0.5)
subplot(2,1,2)
fill_between(x,Vx,x*0,alpha=0.5,color='r')
show()
```
